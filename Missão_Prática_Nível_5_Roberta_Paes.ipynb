```python
#Passo 1: Criando um projeto no Google Colab
#Abri o Google Colab em https://colab.research.google.com/.
#Cliquei em "Novo Notebook" para criar um novo projeto.

#Passo 2: Importei as bibliotecas necessárias
#Para este caso, precisei das seguintes bibliotecas:random, faker, csv, pandas, matplotlib.pyplot e wordcloud

!pip install faker wordcloud matplotlib

#Importei as bibliotecas
import random # 'random' fornece funções para gerar números pseudoaleatórios
import faker # 'faker' permite gerar dados falsos de maneira realista, como nomes, endereços, datas, números de telefone, etc.
import csv #csv' permite ler e escrever dados em formatos tabulares simples, como planilhas.
import pandas as pd #'pandas' fornece estruturas de dados e funções para manipulação e análise eficientes de dados tabulares, como DataFrames
import matplotlib.pyplot as plt #'matplotlib.pyplot' é um submódulo que fornece uma interface para criar gráficos, plotagens e figuras.
from wordcloud import WordCloud # é usada para criar nuvens de palavras a partir de texto,é uma representação visual em que as palavras são exibidas em diferentes tamanhos, com base em sua frequência no texto

#Passo 3: Gerar dados de nome e pontuação
#Criei uma nova célula de código e gerei os dados de nome e pontuação usando a biblioteca "faker" e o módulo "random".
fake = faker.Faker()
num_people = 100
data = [(fake.name(), random.randint(1, 100)) for _ in range(num_people)]

#Passo 4: Gravar os dados em um arquivo de texto
#Criei uma nova célula de código e gravei os dados em um arquivo de texto. Usei a função "open" para fazer isso.
# Salvei os dados em um arquivo de texto (CSV)
with open('dados.txt', 'w') as file:
    for name, score in data:
        file.write(f"{name}: {score}\n")

#Passo 5: Recuperar os dados do arquivo de texto
#Criei uma nova célula de código e recuperei os dados do arquivo de texto que gravei anteriormente:
data = []
with open('dados.txt', 'r') as file:
    for line in file:
        parts = line.strip().split(': ')
        name = parts[0]
        score = int(parts[1])
        data.append((name, score))

#Passo 6: Visualizar as pontuações em um histograma
#Criei uma nova célula de código para criar um histograma das pontuações usando o Matplotlib:
scores = [score for _, score in data]
plt.hist(scores, bins=10, edgecolor='black')
plt.xlabel('Pontuação')
plt.ylabel('Número de Pessoas')
plt.title('Histograma de Pontuações')
plt.show()

#Passo 7: Visualizar as pontuações em uma nuvem de palavras
#Criei uma nova célula de código para criar uma nuvem de palavras das pontuações usando a biblioteca "wordcloud":
text = ' '.join([name for name, _ in data])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nuvem de Palavras das Pontuações')
plt.show()

#Passo 8: Compartilhar o projeto
#Após concluir todas as etapas acima, para compartilhar o projeto do Google Colab siga as etapas:
#Clique no botão "Compartilhar" localizado no canto superior direito da interface do Google Colab.
#Uma caixa de diálogo será exibida, permitindo que você compartilhe o notebook. Nela, você pode definir as configurações de compartilhamento.
#No campo "Link para compartilhar", clique na opção "Qualquer pessoa com o link" (ou uma opção semelhante) para garantir que qualquer pessoa com o link possa acessar o notebook.
#Certifique-se de que a opção "Visualizador" esteja selecionada, o que permitirá que as pessoas vejam o notebook, mas não façam edições.
#Agora, você pode copiar o link gerado na parte inferior da caixa de diálogo.
#Compartilhe esse link com as pessoas que deseja compartilhar o seu trabalho.



```
